```python
import csv
f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)
mx=[0]*24
mx_station=['']*24
for row in data:
    row[4:]=map(int, row[4:])
    for j in range(24):
        b=row[j*2+5]
        if b>mx[j]:
            mx[j]=b
            mx_station[j]=row[3]+'('+str(j+4)+'시)'
print(mx_station,  mx)
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation=90)

plt.show()
```

```python
%matplotlib inline
import csv
import matplotlib.pyplot as plt
f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)
s_in=[0]*24
s_out=[0]*24
for row in data:
    row[4:]=map(int, row[4:])
    for i in range(24):
        s_in[i]+=row[4+i*2]
        s_out[i]+=row[5+i*2]
        
plt.rc('font', family='Malgun GOthic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.xticks(range(24), range(4, 28))
plt.show()
```
